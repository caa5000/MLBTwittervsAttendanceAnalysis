```python
from datetime import date
# date.today()

import datetime
#Create a representation of pull time to get accurate post rate for last 150 posts. 
datetimepull=datetime.datetime(2016, 5, 5, 13, 23, 0)
#Astrospull=Astrostweets[150].created_at #test

#datetimepull-Astrospull #test
```

```python
#How long ago was the 150th tweet (not including re-tweets)? This will give a flavor of RECENT post rate.
TweetRatioData = []
for team in team_tweet_list:
    TweetRatioData.append({'150thTweet':team[150].created_at})
tweetTiming = pd.DataFrame(TweetRatioData)
tweetTiming

#How long ago was that 150th tweet? What is the post rate in a reasonable value? Minutes
import datetime
tweetTiming['TimeSince'] = tweetTiming['150thTweet'].apply(lambda x: datetimepull-x)
tweetTiming['SecondsSince'] = tweetTiming['150thTweet'].apply(lambda x: (datetimepull-x).total_seconds())
tweetTiming['SecondsSince'] = tweetTiming['SecondsSince'].astype(int)
tweetTiming['MinsBetweenTweetsLast150'] = tweetTiming['SecondsSince'].apply(lambda x: x/150/60)
tweetTiming


```
